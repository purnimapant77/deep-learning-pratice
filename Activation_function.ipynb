```python
import math
```

```python
#sigmoid function
def sigmoid(a):
  return 1/(1+math.exp(-a))
```

```python
sigmoid(100)
```

```python
sigmoid(1998)
```

```python
sigmoid(-99)
```

```python
#tanh function
import numpy as np
```

```python
def tanh_activation(x):
  return np.tanh(x)
```

```python
tanh_activation(-99)
```

```python
tanh_activation(1)
```

```python
#ReLU
def relu(x):
  return max(0,x)
```

```python
relu(10)
```

```python
relu(-1)
```

```python
#leaky ReLU
def leaky_relu(x):
  return max(0.1*x,x)
```

```python
leaky_relu(-100)
```
