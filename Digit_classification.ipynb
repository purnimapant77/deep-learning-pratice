```python
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
(X_train,y_train),(X_test,y_test)=keras.datasets.mnist.load_data()
```

```python
print(len(X_train))
print(len(X_test))
```

```python
X_train=X_train/255
X_test=X_test/255
```

```python
y_train[:9]
```

```python
flattened_X_train=X_train.reshape(len(X_train),28*28)
```

```python
flattened_X_train.shape
```

```python
flattened_x_test=X_test.reshape(len(X_test),28*28)
flattened_x_test.shape
```

```python
model=keras.Sequential([
  keras.layers.Dense(10,input_shape=(784,),activation="sigmoid")
])
```

```python
model.compile(
    optimizer='adam',
    loss="sparse_categorical_crossentropy",
    metrics=['accuracy']
)
```

```python
model.fit(flattened_X_train,y_train,epochs=6)
```

```python
test_loss, test_accuracy = model.evaluate(flattened_x_test, y_test)
print("Test accuracy:", test_accuracy)
```
